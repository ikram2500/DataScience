```python
import torch
from transformers import pipeline

classifier = pipeline(task="text-classification", 
                      model="SamLowe/roberta-base-go_emotions", 
                      top_k=None,
                      device = "cuda" if torch.cuda.is_available() else "cpu")

sentences = ["I am not having a great day"]

model_outputs = classifier(sentences)
print(model_outputs[0])
# produces a list of dicts for each of the labels
```

```python
import numpy as np 

emotion_labels = ["netral","sadness","admiration","approval","joy","realization","disappointment","caring","relief","annoyance","optimism","excitement","grief","pride", "desire","nervousness","disapproval","disgust"]

isbn = []
emotion_scores = {label: [] for label in emotion_labels}

def calculate_max_emotion_scores(predictions):
    per_emotion_scores = {label: [] for label in emotion_labels}
    for prediction in predictions:
        sorted_predictions = sorted(prediction, key=lambda x:x['label'])
        for index, label in enumerate(emotion_labels):
            per_emotion_scores[label].append(sorted_predictions[index]['score'])

    return {label: np.max(scores) for label, scores in per_emotion_scores.items()}
```

```python
for i in range(10):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(". ")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
from tqdm import tqdm 


emotion_labels = ["netral","sadness","admiration","approval","joy","realization","disappointment","caring","relief","annoyance","optimism","excitement","grief","pride", "desire","nervousness","disapproval","disgust"]

isbn = []
emotion_scores = {label: [] for label in emotion_labels}
for i in tqdm(range(len(books))):
    isbn.append(books["isbn13"][i])
    sentences = books["description"][i].split(". ")
    predictions = classifier(sentences)
    max_scores = calculate_max_emotion_scores(predictions)
    for label in emotion_labels:
        emotion_scores[label].append(max_scores[label])
```

```python
emotion_df = pd.DataFrame(emotion_scores)
emotion_df["isbn13"] = isbn
```

```python
emotion_df.head(5)
```

```python
books = pd.merge(books, emotion_df, on="isbn13", how="left")
```

```python
books.to_csv("books_with_emotions.csv", index=False)
```

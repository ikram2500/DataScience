### Patterns of Missing data

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().T, cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing values")

plt.title("Patterns of Missing Data in Books Dataset")
plt.show()
```

```python
missing_value_count = books.isna().sum()
print(f"Number of columns with missing values: {missing_value_count}")
```

```python
import numpy as np 

books['missing_description'] = np.where(books['description'].isna(), 1, 0)
books['missing_description'].value_counts().plot(kind='bar')
```

```python
books['age_of_book'] = 2023 - books['published_year']
books['age_of_book'].describe()
```

```python
columns_of_interest = ['num_pages', 'age_of_book', 'missing_description', 'average_rating']

correlation_matrix = books[columns_of_interest].corr(method='spearman')

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', 
                      cbar_kws={"label": "Spearman Correlation"},)
heatmap.set_title('Spearman Correlation')
plt.show()
```

```python
# Identify rows with missing values in key columns
books[(books["description"].isna()) |
         (books["num_pages"].isna()) | 
         (books["average_rating"].isna()) |
         (books["published_year"].isna()) 
         
      ]
```

```python
book_missing = books[~(books["description"].isna()) &
         ~(books["num_pages"].isna()) &
         ~(books["average_rating"].isna()) &
         ~(books["published_year"].isna()) 
         
      ]
```

```python
book_missing
```

```python
book_missing['categories'].value_counts().reset_index().sort_values(by='count', ascending=False)
```

```python
book_missing["words_in_description"] = book_missing["description"].str.split().str.len()
```

```python
book_missing[0:10]
```

```python
# Plot histogram of words_in_description
plt.figure(figsize=(8, 5))
book_missing["words_in_description"].hist(bins=100)
plt.xlabel("Number of Words in Description")
plt.ylabel("Number of Books")
plt.title("Distribution of Words in Book Descriptions")
plt.show()
```

```python
book_missing_25_words = book_missing[book_missing["words_in_description"] >= 25]
```

```python
book_missing_25_words
```

```python
book_missing_25_words["title_and_subtitle"] = (
    np.where(book_missing_25_words["subtitle"].isna(), book_missing_25_words['title'],
             book_missing_25_words[['title','subtitle']].astype(str).agg(": ".join, axis=1))
)
```

```python
book_missing_25_words
```

```python
book_missing_25_words["tagged_description"] = book_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_missing_25_words
```

```python
(
    book_missing_25_words
    .drop(["subtitle","missing_description", "age_of_book", "words_in_description"], axis=1)
    .to_csv("books_cleaned.csv", index=False)
)
```
